```python
import scipy.io as scio
import numpy as np
import matplotlib.pyplot as plt
dataset = scio.loadmat("N09_M07_F10_KA04_1.mat")
data_vib = dataset["N09_M07_F10_KA04_1"]["Y"][0][0][0][6][2][0]
data_vib    # 256001,1      fs 64khz
```

```python
# 振动信号
s_p = data_vib.shape[0]
fs = 64000
s_t = s_p / fs  # 4s
X = np.linspace(0,s_t,s_p)
Y = data_vib.flatten()
plt.figure(figsize=(15,3))
plt.title("vibration signal")
plt.plot(X, Y)
plt.tight_layout()
```

```python
# 电流信号--current signal  

o_data_cur_1 = dataset["N09_M07_F10_KA04_1"]["Y"][0][0][0][1][2][0]
o_data_cur_2 = dataset["N09_M07_F10_KA04_1"]["Y"][0][0][0][2][2][0]
data_cur_1 = o_data_cur_1.flatten()
data_cur_2 = o_data_cur_2.flatten()

s_p_cur = len(data_cur_1) # 256001
fs_cur = 64000
s_t_cur = s_p_cur / fs_cur  # 4.s

X = np.linspace(0,s_t_cur, s_p_cur)

fig_cur = plt.figure(figsize=(15,6))
plt.subplot(2,1,1)
plt.plot(X,data_cur_1)
plt.title("current signal 1")
plt.subplot(2,1,2)
plt.plot(X, data_cur_2)
plt.title("current signal 2")
plt.tight_layout()
```
